## Ticker Module
To access data for a single stock

```python
import yfinance as yf
stock = yf.Ticker('META')
```

```python
stock.info
```

## Price History

```python
harga = stock.history(period='3mo')
harga.head(3)
```

```python
harga['Returns'] = harga['Close'].pct_change()
harga.head()
```

```python
display(stock.history(period='1d'))
```
